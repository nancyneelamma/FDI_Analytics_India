**KEY METRICS FOR FDI ANALYSIS**

**Sector-Wise FDI Analysis**

```python
# Change rows to years

data1 = df.melt(id_vars=['Sector'], var_name='Year', value_name='Average FDI')

#Calculate fdi for all the sectors

sectors_fdi = data1.groupby('Sector')['Average FDI'].mean()
print(sectors_fdi)

# Plotting bar graph for Sector-Wise Distribution

plt.figure(figsize=(16, 12))
sns.barplot(x=sectors_fdi.values,y=sectors_fdi.index,color='SkyBlue')
plt.title('Sector Analysis of Foreign Direct Investments')
plt.xlabel('Average FDI')
plt.ylabel('Sectors')
plt.tight_layout()
plt.show()
```

**Year-Wise FDI Analysis**

```python
#Calculate each year total FDI
sectors=df['Sector']
year_total_FDI = df.drop(columns='Sector').sum()
year_average_FDI = df.drop(columns='Sector').mean()
years=pd.DataFrame({'Years':year_average_FDI.index,'Total FDI':year_total_FDI.values,
                    'Average FDI':year_average_FDI.values})
print(years)

# Plotting line graph for Year-Wise Distribution

plt.figure(figsize=(12, 6))
sns.lineplot(data=year_average_FDI,marker='o')
plt.title('Yearly Analysis of Foreign Direct Investments')
plt.xlabel('Year')
plt.ylabel('Average FDI')
plt.tight_layout()
plt.grid(True)
plt.show()
```

**Yearly Growth Rate Analysis**

```python
# Change rows to years

data2 = df.melt(id_vars=['Sector'],var_name='Year',value_name='Total FDI')

#Calculate fdi across all sectors for each year

yearly_fdi = data2.groupby('Year')['Total FDI'].sum()
# Calculate Growth Rate for each year

yearly_growth = yearly_fdi.pct_change()*100
yearly_growth_data = pd.DataFrame({'Year':yearly_growth.index,'Growth Percentage':yearly_growth.values})
print(yearly_growth_data)
# Plotting line graph

plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_growth.index,y=yearly_growth.values,marker='o',color='Purple')
plt.title('Growth Rate by Years')
plt.xlabel('Year')
plt.ylabel('Growth Rate(%)')
plt.tight_layout()
plt.grid(True)
plt.show()
```

**Summary**

**Insights:**

**KEY METRICS FOR FDI ANALYSIS**

**Sector-Wise FDI Analysis**

*  From the above bar plots we can see that Services Sector has consitently making investments every year and it has made it's highest investments in the year 2016-2017. On an average if we consider for  the services sector has been has always been one the top inverstors for majority years.
*  There are some sectors that have made quite good investments each year and has again been consistent with doing so, here are some of the sectors: metallurgical industries, mining, power, computer software and hardware, telecommunications.
*  We can also see some of the sectors have made very few investments or  has zero investments, if we consider on an average the coal production has majorly made either zero investments or zero inverstments for majority of the years.
* And after sector Coal Production Coir and Ports have fewer investments or no investments in some years.


**Year-Wise FDI Analysis**

* We can observe here the investments starts from year 2000-01 with (fdi=2378.71) and has a slight increase till year 2002-03 with (fdi=2704.32) have seen growth and consistency, also the was a slight drop in the investments made in year 2003-04 with (fdi=2187.85). 
* From year 2004-05 with (fdi=3218.69) there is steady increase in the investments and year          2008-09 with (fdi=31395.96) has been a breakthrough, for the sectors but then there was a drop down from year 2009-2010 with (fdi=325834.38) until year 2010-2011 with (fdi=21383.07).
* Again there was another breakthrough i.e., the sectors have witnessed a rise in the investment in the year 2011-12 with (fdi=35120.78) but year 2012-13 faced decreased in investments with (fdi=22423.59)
* Lastly, year 2013-14(fdi=24299.32) to year 2016-17(fdi=43478.26) the sectors have seen huge increase in the investments made from the sectors.
